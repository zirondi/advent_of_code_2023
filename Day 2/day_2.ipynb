# Part One

```python
import re

regex_pattern = r'\d+'

bag_contents = {
    'red': 12,
    'green': 13,
    'blue': 14,
}

input_path = 'input.txt'
with open(input_path, 'r') as f:
    data = f.read().split('\n')


possible_games_id_sum = 0


for line in data:

    splitted_line = line.split(':')

    game_id = int(re.search(regex_pattern, splitted_line[0]).group())

    turns = splitted_line[1].split(';')
    
    game_possible = True


    for turn in turns:
        splitted_turn = turn.split(',')

        colors = turn.split(',')

        for color in colors:

            color = color.strip()
            splited_color = color.split(' ')

            qtd = int(splited_color[0])
            color = splited_color[-1]

            if qtd > bag_contents[color]:
                game_possible = False

                break	

        if game_possible is False:
            break


    if game_possible is True:
        possible_games_id_sum += game_id



possible_games_id_sum
```

# Part Two

```python
from math import prod

cubes_color = ['red', 'green', 'blue']

input_path = 'input.txt'
with open(input_path, 'r') as f:
    data = f.read().split('\n')


fewest_possible_power_sum = 0

for line in data:

    splitted_line = line.split(':')

    turns = splitted_line[1].split(';')
    

    min_value_possible = {key:float('-inf') for key in cubes_color}

    for turn in turns:
        splitted_turn = turn.split(',')

        colors = turn.split(',')

        for color in colors:

            color = color.strip()
            splited_color = color.split(' ')

            qtd = int(splited_color[0])
            color = splited_color[-1]

            if qtd > min_value_possible[color]:
                min_value_possible[color] = qtd
    
    min_valu_prod = prod(min_value_possible.values())
    fewest_possible_power_sum += min_valu_prod


fewest_possible_power_sum
```
